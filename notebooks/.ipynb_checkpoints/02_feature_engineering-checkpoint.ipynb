# üîß Feature Engineering for Upsell Prediction
## AI Upsell Prediction System - Phase 2

**Objective**: Transform raw customer data into ML-ready features

Based on EDA findings:
- 40,729 duplicate rows need cleaning
- Clear usage patterns for feature creation
- 10.4% churn rate provides good target balance

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
import warnings
warnings.filterwarnings('ignore')

print("üîß Feature Engineering Phase Started!")
print("üìä Libraries loaded successfully!")
```

```python
# Load the raw data
df_raw = pd.read_csv('../data/raw/telecom_data.csv', low_memory=False)

print(f"üìä Raw data loaded: {df_raw.shape}")
print(f"üë• Total customers: {df_raw.shape[0]:,}")
print(f"üìã Original features: {df_raw.shape[1]}")

print("\nüîç Data Overview:")
display(df_raw.head())
```

```python
# Data cleaning - Remove duplicates and handle missing values
print("üßπ Starting data cleaning...")
print(f"üìä Original shape: {df_raw.shape}")

# Remove duplicates
initial_duplicates = df_raw.duplicated().sum()
df_clean = df_raw.drop_duplicates()
print(f"üóëÔ∏è Removed {initial_duplicates:,} duplicate rows")

# Handle missing values in numerical columns
missing_before = df_clean.isnull().sum().sum()
numerical_cols = df_clean.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    if df_clean[col].isnull().any():
        df_clean[col].fillna(df_clean[col].median(), inplace=True)

# IMPORTANT FIX: Handle missing values in 'Churn' column specifically
# If 'Churn' is an object type with NaN, or a boolean with NaN, fill with False (0)
if 'Churn' in df_clean.columns and df_clean['Churn'].isnull().any():
    print("üîß Filling NaN in 'Churn' column with False...")
    df_clean['Churn'].fillna(False, inplace=True)

missing_after = df_clean.isnull().sum().sum()
print(f"üîß Fixed {missing_before - missing_after} missing values")

print(f"‚úÖ Clean data shape: {df_clean.shape}")
print(f"üìà Data quality improved: {((df_raw.shape[0] - df_clean.shape[0]) / df_raw.shape[0] * 100):.1f}% duplicates removed")
```

```python
# Feature Engineering - Create new features
print("‚öôÔ∏è Creating engineered features...")

df_features = df_clean.copy()

# 1. Usage aggregation features
print("üìä Creating usage aggregation features...")
df_features['Total_Minutes'] = df_features['Day Mins'] + df_features['Eve Mins'] + df_features['Night Mins']
df_features['Total_Calls'] = df_features['Day Calls'] + df_features['Eve Calls'] + df_features['Night Calls']
df_features['Total_Charges'] = df_features['Day Charge'] + df_features['Eve Charge'] + df_features['Night Charge'] + df_features['Intl Charge']

# 2. Usage intensity features
print("üìà Creating usage intensity features...")
df_features['Avg_Call_Duration'] = df_features['Total_Minutes'] / (df_features['Total_Calls'] + 1)
df_features['Day_Call_Duration'] = df_features['Day Mins'] / (df_features['Day Calls'] + 1)
df_features['Eve_Call_Duration'] = df_features['Eve Mins'] / (df_features['Eve Calls'] + 1)
df_features['Night_Call_Duration'] = df_features['Night Mins'] / (df_features['Night Calls'] + 1)

# 3. Usage pattern features (ratios)
print("üéØ Creating usage pattern features...")
df_features['Day_Usage_Ratio'] = df_features['Day Mins'] / (df_features['Total_Minutes'] + 1)
df_features['Eve_Usage_Ratio'] = df_features['Eve Mins'] / (df_features['Total_Minutes'] + 1)
df_features['Night_Usage_Ratio'] = df_features['Night Mins'] / (df_features['Total_Minutes'] + 1)
df_features['Intl_Usage_Ratio'] = df_features['Intl Mins'] / (df_features['Total_Minutes'] + 1)

# 4. Revenue-based features
print("üí∞ Creating revenue-based features...")
df_features['Revenue_Per_Minute'] = df_features['Total_Charges'] / (df_features['Total_Minutes'] + 1)
df_features['Day_Revenue_Rate'] = df_features['Day Charge'] / (df_features['Day Mins'] + 1)
df_features['Eve_Revenue_Rate'] = df_features['Eve Charge'] / (df_features['Eve Mins'] + 1)
df_features['Night_Revenue_Rate'] = df_features['Night Charge'] / (df_features['Night Mins'] + 1)

# 5. Customer lifecycle features
print("üìÖ Creating customer lifecycle features...")
df_features['Account_Length_Months'] = df_features['Account Length'] / 30.44
df_features['Usage_Per_Day'] = df_features['Total_Minutes'] / (df_features['Account Length'] + 1)
df_features['Revenue_Per_Day'] = df_features['Total_Charges'] / (df_features['Account Length'] + 1)
df_features['Service_Calls_Per_Month'] = df_features['CustServ Calls'] / (df_features['Account_Length_Months'] + 1)

# 6. Behavioral indicators (binary features)
print("üé≠ Creating behavioral indicators...")
df_features['Is_Heavy_Day_User'] = (df_features['Day Mins'] > df_features['Day Mins'].quantile(0.75)).astype(int)
df_features['Is_Heavy_Eve_User'] = (df_features['Eve Mins'] > df_features['Eve Mins'].quantile(0.75)).astype(int)
df_features['Is_Heavy_Night_User'] = (df_features['Night Mins'] > df_features['Night Mins'].quantile(0.75)).astype(int)
df_features['Is_Intl_User'] = (df_features['Intl Calls'] > 0).astype(int)
df_features['Is_High_Service_User'] = (df_features['CustServ Calls'] > 2).astype(int)
df_features['Has_Voicemail'] = (df_features['VMail Message'] > 0).astype(int)
df_features['Is_High_Value_Customer'] = (df_features['Total_Charges'] > df_features['Total_Charges'].quantile(0.8)).astype(int)

# 7. Risk indicators
print("‚ö†Ô∏è Creating risk indicators...")
df_features['Churn_Risk_Score'] = (
    (df_features['CustServ Calls'] > 3).astype(int) * 3 +
    (df_features['Total_Minutes'] < df_features['Total_Minutes'].quantile(0.2)).astype(int) * 2 +
    (df_features['Account Length'] < 90).astype(int) * 1
)

# Handle infinite values
df_features = df_features.replace([np.inf, -np.inf], np.nan)
numerical_features = df_features.select_dtypes(include=[np.number]).columns
for col in numerical_features:
    if df_features[col].isnull().any():
        df_features[col].fillna(df_features[col].median(), inplace=True)

print(f"‚úÖ Feature engineering complete!")
print(f"üìä New shape: {df_features.shape}")
print(f"üÜï Features added: {df_features.shape[1] - df_raw.shape[1]}")
```

```python
# Analyze the new features
new_features = [col for col in df_features.columns if col not in df_raw.columns]
print(f"\nüÜï New features created ({len(new_features)}):")
for i, feature in enumerate(new_features, 1):
    print(f"{i:2d}. {feature}")

print(f"\nüìà Feature engineering summary:")
print(f"   Original features: {df_raw.shape[1]}")
print(f"   Engineered features: {len(new_features)}")
print(f"   Total features: {df_features.shape[1]}")
print(f"   Customers after cleaning: {df_features.shape[0]:,}")
```

```python
# Key feature statistics
key_features = ['Total_Minutes', 'Total_Charges', 'Avg_Call_Duration', 'Revenue_Per_Minute', 
                'Day_Usage_Ratio', 'Service_Calls_Per_Month', 'Churn_Risk_Score']

print("üìä Key Engineered Features Summary:")
display(df_features[key_features].describe())
```

```python
# Visualize engineered features
plt.figure(figsize=(20, 12))

# Total usage distribution
plt.subplot(3, 4, 1)
plt.hist(df_features['Total_Minutes'], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Total Minutes Distribution', fontweight='bold')
plt.xlabel('Total Minutes')

# Revenue per minute
plt.subplot(3, 4, 2)
plt.hist(df_features['Revenue_Per_Minute'], bins=50, alpha=0.7, color='orange', edgecolor='black')
plt.title('Revenue Per Minute Distribution', fontweight='bold')
plt.xlabel('Revenue Per Minute')

# Usage ratios
plt.subplot(3, 4, 3)
usage_ratios = ['Day_Usage_Ratio', 'Eve_Usage_Ratio', 'Night_Usage_Ratio']
df_features[usage_ratios].boxplot(ax=plt.gca())
plt.title('Usage Ratios by Time Period', fontweight='bold')
plt.xticks(rotation=45)

# Churn risk score
plt.subplot(3, 4, 4)
risk_counts = df_features['Churn_Risk_Score'].value_counts().sort_index()
plt.bar(risk_counts.index, risk_counts.values, alpha=0.7, color='red')
plt.title('Churn Risk Score Distribution', fontweight='bold')
plt.xlabel('Risk Score')

# Binary features
binary_features = ['Is_Heavy_Day_User', 'Is_Heavy_Eve_User', 'Is_Intl_User', 'Is_High_Service_User']
for i, feature in enumerate(binary_features, 5):
    plt.subplot(3, 4, i)
    feature_counts = df_features[feature].value_counts()
    plt.pie(feature_counts.values, labels=['No', 'Yes'], autopct='%1.1f%%')
    plt.title(f'{feature.replace("_", " ")}', fontweight='bold')

# Average call duration
plt.subplot(3, 4, 9)
plt.hist(df_features['Avg_Call_Duration'], bins=50, alpha=0.7, color='purple', edgecolor='black')
plt.title('Average Call Duration', fontweight='bold')
plt.xlabel('Minutes per Call')

# Service calls per month
plt.subplot(3, 4, 10)
plt.hist(df_features['Service_Calls_Per_Month'], bins=30, alpha=0.7, color='green', edgecolor='black')
plt.title('Service Calls Per Month', fontweight='bold')
plt.xlabel('Calls per Month')

# Usage per day
plt.subplot(3, 4, 11)
plt.hist(df_features['Usage_Per_Day'], bins=50, alpha=0.7, color='brown', edgecolor='black')
plt.title('Usage Per Day', fontweight='bold')
plt.xlabel('Minutes per Day')

# Revenue per day
plt.subplot(3, 4, 12)
plt.hist(df_features['Revenue_Per_Day'], bins=50, alpha=0.7, color='pink', edgecolor='black')
plt.title('Revenue Per Day', fontweight='bold')
plt.xlabel('Dollars per Day')

plt.tight_layout()
plt.show()

print("\nüéØ Feature Engineering Insights:")
print(f"üìû Heavy day users: {df_features['Is_Heavy_Day_User'].sum():,} customers ({df_features['Is_Heavy_Day_User'].mean():.1%})")
print(f"üåÜ Heavy evening users: {df_features['Is_Heavy_Eve_User'].sum():,} customers ({df_features['Is_Heavy_Eve_User'].mean():.1%})")
print(f"üåç International users: {df_features['Is_Intl_User'].sum():,} customers ({df_features['Is_Intl_User'].mean():.1%})")
print(f"üìû High service users: {df_features['Is_High_Service_User'].sum():,} customers ({df_features['Is_High_Service_User'].mean():.1%})")
print(f"üí∞ High value customers: {df_features['Is_High_Value_Customer'].sum():,} customers ({df_features['Is_High_Value_Customer'].mean():.1%})")
```

```python
# Prepare data for machine learning
print("üéØ Preparing data for machine learning...")

# Convert Churn to binary (True/False to 1/0)
# Ensure Churn is boolean before converting to int
df_features['Churn'] = df_features['Churn'].astype(bool)
df_features['Churn_Binary'] = df_features['Churn'].astype(int)

# Select features for ML (exclude identifier and original target columns)
exclude_columns = ['Phone Number', 'Churn']
feature_columns = [col for col in df_features.columns if col not in exclude_columns]

# Separate features and target
X = df_features[feature_columns]
y = df_features['Churn_Binary']

print(f"‚úÖ ML data preparation complete!")
print(f"üìä Features for ML: {X.shape[1]}")
print(f"üë• Samples: {X.shape[0]:,}")
print(f"üéØ Target distribution: {y.value_counts().to_dict()}")
print(f"üìà Positive class rate: {y.mean():.1%}")
```

```python
# Save processed data
print("üíæ Saving processed data...")

# Create processed data directory if it doesn't exist
import os
os.makedirs('../data/processed', exist_ok=True)

# Save the complete engineered dataset
df_features.to_csv('../data/processed/engineered_features.csv', index=False)
print(f"‚úÖ Saved engineered dataset: {df_features.shape}")

# Save ML-ready data
X.to_csv('../data/processed/X_features.csv', index=False)
y.to_csv('../data/processed/y_target.csv', index=False)
print(f"‚úÖ Saved ML-ready features: {X.shape}")
print(f"‚úÖ Saved target variable: {y.shape}")

# Save feature names
pd.Series(feature_columns).to_csv('../data/processed/feature_names.csv', index=False, header=['feature_name'])
print(f"‚úÖ Saved feature names: {len(feature_columns)} features")

print("\nüéâ Feature Engineering Phase Complete!")
print("üöÄ Ready for Model Development Phase!")
```
